# Gradient Boosting

Gradient boosting is a form of Ensemble Learning
* Ensemble ~ "together" in French

work on Below principle:

" Build up your weaknesses until they become your strong points"

        Many machine learning models come together to work on the training data.
        Many weak learners will make strong learners.
        
 Focuses on what previous model failed to learn
 
 Weak Learners are called "Ensemble" of learners.
 In practice: 100-200 weak learners, each learning from previous mistakes.
 
         Gradient Boosting will not work if the weak learners use MSE regression.
         Regression using Decision Trees work well.
         
 Decision Tree : 
 
         Non-parametric ML technique
         Hyperparameter:how many levels?
         
 * Gradient Boosting works best with a large number of weak learners
 Large ~ several hundred
 Early learners learn the most
 Later learners learn from the mistakes of early learners
 
 
 
 Shrinkage Factor :
      
      * Scale output of each model by a constant factor.
      * High shrinkage factor scales down importance of each learner
      * Slow down learning, reduces overfitting
      * More learners ~ shrink a lot
      * Few learners ~ shrink a little
      * Typical values ~ 0.1,1

     GradientBoostingRegressor(max_depth = 4, n_estimators=200, learning rate = 0.1)

     * max_depth = maximum depth of each individual decision tree

```python
# coverting categorical value to one hot representation, because all these features are important for model learning
auto_data = pd.get_dummies(auto_data,columns=['make','fuel-type', 'aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','fuel-system'])
auto_data.head()
```

```python
# spliting data into training and testing set
from sklearn.model_selection import train_test_split
X = auto_data.drop('price', axis =1)

# Taking the labels (price)
Y = auto_data['price']

X_train, x_test, Y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

# Estimator for GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor

params = {'n_estimators': 500,'max_depth':6 , 'min_samples_split':2,'learning_rate': 0.01, 'loss':'ls'}
# min_samples_split - node split , ls - least squares regression
gbr_model = GradientBoostingRegressor(**params)
gbr_model.fit(X_train, Y_train)
```

```python
gbr_model.score(X_train,Y_train)
```

```python
y_predict = gbr_model.predict(x_test)
```

```python
r_square = gbr_model.score(x_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

gbr_model_mse = mean_squared_error(y_predict, y_test)
gbr_model_mse
```

```python
import math
math.sqrt(gbr_model_mse)
```
